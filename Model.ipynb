```python
predictors = ["Age", "G", "GS", "MP", "FG", "FGA", 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'W', 'L', 'W/L%',
       'GB', 'PS/G', 'PA/G', 'SRS']
```

```python
train = stats[~(stats["Year"] == 2018)]
test = stats[stats["Year"] == 2018]
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=.1)
```

```python
reg.fit(train[predictors],train["Share"])
```

```python
predictions = reg.predict(test[predictors])
predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
```

```python
combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
combination.sort_values("Share", ascending=False).head(20)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(combination["Share"], combination["predictions"])
```

```python
actual = combination.sort_values("Share", ascending=False)
predicted = combination.sort_values("predictions", ascending=False)
actual["Rk"] = list(range(1,actual.shape[0]+1))
predicted["Predicted_Rk"] = list(range(1,predicted.shape[0]+1))
```

```python
def find_ap(combination):
    real = combination.sort_values("Share", ascending=False).head(5)
    calc = combination.sort_values("predictions", ascending=False)
    ps = []
    found = 0
    count = 1
    for index,row in calc.iterrows():
        if row["Player"] in real["Player"].values:
            found += 1
            ps.append(found / count)
        count += 1

    return sum(ps) / len(ps)
```

```python
ap = find_ap(combination)
ap
```

```python
years = list(range(1991,2021))
aps = []
all_calculated_values = []
for year in years[5:]:
    training_values = stats[stats["Year"] < year]
    testing_values = stats[stats["Year"] == year]
    reg.fit(training_values[predictors],training_values["Share"])
    predictions = reg.predict(testing_values[predictors])
    predictions = pd.DataFrame(predictions, columns=["predictions"], index=testing_values.index)
    val = pd.concat([testing_values[["Player", "Share"]], predictions], axis=1)
    all_calculated_values.append(val)
    aps.append(find_ap(val))
```

```python
sum(aps) / len(aps)

```

```python
def add_ranks(calculated_values):
    calculated_values = calculated_values.sort_values("predictions", ascending=False)
    calculated_values["Predicted_Rk"] = list(range(1,calculated_values.shape[0]+1))
    calculated_values = calculated_values.sort_values("Share", ascending=False)
    calculated_values["Rk"] = list(range(1,calculated_values.shape[0]+1))
    calculated_values["Diff"] = (calculated_values["Rk"] - calculated_values["Predicted_Rk"])
    return calculated_values
add_ranks(all_calculated_values[1])
```

```python
def backtest(stats, model, years, predictors):
    aps = []
    all_calculated_values = []
    for year in years:
        train = stats[stats["Year"] < year]
        test = stats[stats["Year"] == year]
        model.fit(train[predictors],train["Share"])
        predictions = model.predict(test[predictors])
        predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
        combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
        combination = add_ranks(combination)
        all_calculated_values.append(combination)
        aps.append(find_ap(combination))
    return sum(aps) / len(aps), aps, pd.concat(all_calculated_values)
```
