```python
options=webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

driver=webdriver.Chrome(options=options)
```

```python
os.environ["LANGCHAIN_HANDLER"] = "langchain"
```

```python
for year in val_range:
    player_stats_url = "https://www.basketball-reference.com/leagues/NBA_1991_per_game.html"

    url = player_stats_url.format(year)

    driver.get(player_stats_url)
    driver.execute_script("window.scrollTo(1,10000)")
    time.sleep(15)
    file_path=os.path.join('/content/mvp/player_stat',str(year)+'.html')
    with open(file_path, "w+") as f:
        f.write(driver.page_source)
```

```python
dfs = []
for year in val_range:
    file_path=os.path.join('/content/mvp/player_stat',str(year)+'.html')
    with open(file_path) as f:
        page = f.read()

    soup = BeautifulSoup(page, 'html.parser')
    soup.find('tr', class_="thead").decompose()
    player_table = soup.find_all(id="per_game_stats")[0]
    player_df = pd.read_html(str(player_table))[0]
    player_df["Year"] = year
    dfs.append(player_df)
```

```python
players = pd.concat(dfs)
```

```python
team_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_standings.html"
```

```python

for year in val_range:
    url = team_stats_url.format(year)

    data = requests.get(url)
    file_path=os.path.join('/content/mvp/nbadata',str(year)+'.html')
    with open(file_path, "w+") as f:
        f.write(data.text)
```

```python
dfs = []
for year in val_range:
    file_path=os.path.join('/content/mvp/nbadata',str(year)+'.html')
    with open(file_path) as f:
        page = f.read()

    soup = BeautifulSoup(page, 'html.parser')
    soup.find('tr', class_="thead").decompose()
    df2 = soup.find_all(id="divs_standings_E")[0]
    df1 = pd.read_html(str(df2))[0]
    df1["Year"] = year
    df1["Team"] = df1["Eastern Conference"]
    del df1["Eastern Conference"]
    dfs.append(df1)

    table = soup.find_all(id="divs_standings_W")[0]
    df = pd.read_html(str(table))[0]
    df["Year"] = year
    df["Team"] = df["Western Conference"]
    del df["Western Conference"]
    dfs.append(df)
```

```python
train=pd.concat(dfs)
```

```python
train.tail()
```

```python
train.to_csv("dataset.csv")
```
